```python
from Haptik import Haptik
from nltk.tokenize import RegexpTokenizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
path = '~/Github/Haptik_dataset_NLP/data'
x = Haptik(path)
x.wordcloud()
```

```python
tokenizer = RegexpTokenizer(r'\w+')
vect = TfidfVectorizer(stop_words = 'english', ngram_range = (1,3),
        tokenizer = tokenizer.tokenize, min_df = 1, max_df = 0.70, norm = 'l1',
        max_features = 25000, sublinear_tf = True)
model = MultinomialNB(alpha=0.25, class_prior=None, fit_prior=False)
x.classify(vect = vect, model = model)
```
